```python
import pandas as pd
ChinaBank=pd.read_csv('ChinaBank.csv',\
                     index_col='Date')
ChinaBank=ChinaBank.iloc[:,1:]#delete the sequence no.
ChinaBank.index=pd.to_datetime(ChinaBank.index)
Close=ChinaBank.Close

import matplotlib.pyplot as plt
plt.plot(Close['2014'])
```

```python
#add additional information to this picture
plt.plot([1,1,0,0,-1,0,1,1,-1])
plt.rcParams['axes.unicode_minus']=False
plt.xlim(0,8)
plt.ylim(-1,1)
```

```python
plt.plot([1,1,0,0,-1,0,1,1,-1])
plt.ylim(-1.5,1.5)
plt.xticks(range(9),\
          ['2015-02-01','2015-02-02','2015-02-03','2015-02-04','2015-02-05',\
           '2015-02-06','2015-02-07','2015-02-08','2015-02-09'],\
          rotation=45)
```

```python
#add some texts, labels for axes and grids
plt.plot(Close['2014'])
plt.title('''China Bank's Close Price Curve of 2014''',\
         loc='right')
#if we want write Chinese in the plot, we have to use 'rcParams'
plt.xlabel('Dates')
plt.ylabel('Close Price')
plt.grid(True,axis='y')
```

```python
#add more pictures
Open=ChinaBank.Open
plt.plot(Close['2014'],label='Close')
plt.plot(Open['2014'],label='Open')
plt.legend()
```

```python
#other styles of lines, colors
plt.plot(Close['2014'],c='r',label='Close',linestyle='solid')
plt.plot(Open['2014'],c='b',label='Open',ls='-.')
plt.legend()
plt.title('''China Bank's Close and Open Prices Curve of 2014''',\
         loc='right')
plt.xlabel('Dates')
plt.ylabel('Price')
plt.grid(True,axis='y')
```

```python
#types of points
plt.plot(Close['2015'],marker='o',label='Close')
plt.plot(Open['2015'],marker='*',label='Open')
plt.legend(loc='best')
plt.title('''China Bank's Close and Open Prices Curve of 2015''')
plt.xlabel('Dates')
plt.ylabel('Price')
plt.xticks(rotation=60)
plt.grid(True,axis='both')
```

```python
plt.plot(Close['2015'],'--rD',label='Close')
plt.plot(Open['2015'],'--b',label='Open')
plt.legend(loc='best')
plt.title('''China Bank's Close and Open Prices Curve of 2015''')
plt.xlabel('Dates')
plt.ylabel('Price')
plt.xticks(rotation=90)
plt.grid(True,axis='both')
```

```python
plt.plot(Close['2015'],'--rD',label='Close',linewidth=2)
plt.plot(Open['2015'],'--b',label='Open',lw=10)
plt.legend(loc='best')
plt.title('''China Bank's Close and Open Prices Curve of 2015''')
plt.xlabel('Dates')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.grid(True,axis='both')
```

```python
#Bar charts, which reflect frequency
Close.describe()
a=[0,0,0,0]
for i in Close:
    if (i>2)&(i<=3):
        a[0]+=1
    elif (i>3)&(i<=4):
        a[1]+=1
    elif (i>4)&(i<=5):
        a[2]+=1
    else:
        a[3]+=1
plt.bar([2,3,4,5],a)
```

```python
plt.bar(x=[2,3,4,5],height=a,width=1.0,\
       bottom=0,color='red',edgecolor='k')
plt.title('China Bank Close Price Bar Chart')
```

```python
plt.barh([2,3,4,5],a,height=1.0,color='red',edgecolor='k')
plt.title('China Bank Close Price Bar Chart')
```

```python
#histogram
plt.hist(Close,bins=12)
plt.title('China Bank Close Price Histogram')
```

```python
plt.hist(Close,range=(2.3,5.5),\
        orientation='horizontal',\
        color='r',edgecolor='k')
plt.title('China Bank Close Price Histogram')
```

```python
#histogram w.r.t. cumulative distribution
plt.hist(Close,range=(2.3,5.5),\
        orientation='vertical',\
        cumulative=True,histtype='stepfilled',color='r',edgecolor='k')
plt.title('China Bank Close Price Histogram')
```

```python
#Pie Plot
plt.pie(a,labels=('(2,3]','(3,4]','(4,5]','(5,6]'),\
       colors=('b','g','r','c'),shadow=True)
plt.title('China Bank Close Price Pie Plot')
```

```python
#Box Plot
import numpy as np
prcData=ChinaBank.iloc[:,:4]
data=np.array(prcData)
plt.boxplot(data,labels=('Open','High','Low','Close'))
plt.title('China Bank Close Price Box Plot')
```

```python
#Figures and Axes, which is object-oriented ploting
fig=plt.figure()#construct a new 'white paper'
ax1=fig.add_axes([0.1,0.1,0.3,0.3])
ax2=fig.add_axes([0.5,0.5,0.4,0.4])#two different coordinates on one paper
ax1.plot(Close[:10])
ax2.plot(Open[:10])
ax1.set_title('A')
ax2.set_title('B')
```

```python
#multiplot
Close15=Close['2015']
ax1=plt.subplot(211)
ax1.plot(Close15,color='k')
ax1.set_ylabel('Close')
ax1.set_title('A')

Volume15=ChinaBank.Volume['2015']
Open15=Open['2015']
ax2=plt.subplot(212)
left1=Volume15.index[Close15>=Open15]
hight1=Volume15[left1]
ax2.bar(left1,hight1,color='r')
left2=Volume15.index[Close15<Open15]
hight2=Volume15[left2]
ax2.bar(left2,hight2,color='g')
ax2.set_title('B')
```

```python
#plot more than one curve within one picture
High15=ChinaBank.High['2015']
Low15=ChinaBank.Low['2015']

fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])

ax.plot(Close15,label='Close')
ax.plot(Open15,'--*',label='Open')
ax.plot(High15,'-+',label='Hig')
ax.plot(Low15,'-.>',label='Low')

ax.set_title('China Bank 2015 Prices')
ax.set_ylabel('Prices')
ax.set_xticklabels(Close.index[:10],rotation=45)
ax.legend(loc='best')
```
